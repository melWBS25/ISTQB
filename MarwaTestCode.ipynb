```python
class Calculator:
    """Eine einfache Taschenrechner-Klasse"""

    def addiere(self, a,b):
        """Addiert zwei Zahlen"""
        return a + b
    
    def subtrahiere(self, a, b):
        """subtrahiert zwei Zahlen"""
        return a - b
    def divid(self, a, b):
        """divid zwei Zahlen"""
        if (b==0):
          raise ValueError("b can not be 0") 
        
        return a / b
    
```
